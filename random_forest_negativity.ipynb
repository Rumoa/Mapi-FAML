# Confusion matrix

```python
from sklearn.metrics import confusion_matrix

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import joblib
df = joblib.load("pure_negativity.job")

# clf = RandomForestClassifier(random_state=1)

X = df.drop(columns="negativity")
y = df['negativity']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)



from sklearn.multiclass import OneVsOneClassifier
from sklearn.svm import SVC
clf_svm_ovo = OneVsOneClassifier(
    SVC(random_state=0))
# np.sum(y_test == clf.predict(X_test))/len(y_test)
clf_svm_ovo.fit(X_train, y_train)
results_svm_ovo = clf_svm_ovo.predict(X_test)

confusion_matrix(y_test, results_svm_ovo)
```

```python
from sklearn.metrics import confusion_matrix

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import joblib
df = joblib.load("pure_negativity.job")

# clf = RandomForestClassifier(random_state=1)

X = df.drop(columns="negativity")
y = df['negativity']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)



from sklearn.multiclass import OneVsOneClassifier
from sklearn.svm import SVC
clf_rf_ovo = OneVsOneClassifier(
    RandomForestClassifier(random_state=0))
# np.sum(y_test == clf.predict(X_test))/len(y_test)
clf_rf_ovo.fit(X_train, y_train)
results_rf_ovo = clf_rf_ovo.predict(X_test)

confusion_matrix(y_test, results_rf_ovo)
```

```python

import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(y_test, results_svm_ovo, labels=clf_svm_ovo.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=clf_svm_ovo.classes_)
disp.plot()
plt.title("Confusion matrix for classification of negativity classes \nSVM OVO (80/20)")

plt.savefig("img/svm_conf.jpeg", dpi=300)
plt.savefig("img/svm_conf.pdf", dpi=300)
plt.show()
```
