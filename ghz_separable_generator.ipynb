```python
import qutip as qu
import numpy as np
seed = 1
np.random.seed(seed=seed)
```

```python
import numpy as np
from scipy.constants import pi
```

```python
from qutip.tensor import tensor
from qutip.states import *
from qutip.operators import *
final_states = []
for i in range(1000):
    final_states.append(qu.tensor([qu.rand_ket(2), qu.rand_ket(2), qu.rand_ket(2)]))
```

```python
def rand_ghz():
    delta = np.random.uniform(low=0, high=pi/4 )
    alpha = np.random.uniform(low=0, high=pi/2 )
    beta = np.random.uniform(low=0, high=pi/2 )
    gamma = np.random.uniform(low=0, high=pi/2 )
    phi = np.random.uniform(low=0, high=2*pi )
    K = 1/(1 + np.cos(delta)*np.sin(delta)*np.cos(alpha)*np.cos(beta)*np.cos(phi))

    phi_a = np.cos(alpha)*basis(2,0) + np.sin(alpha)*basis(2,1)
    phi_b = np.cos(beta)*basis(2,0) + np.sin(beta)*basis(2,1)
    phi_c = np.cos(gamma)*basis(2,0) + np.sin(gamma)*basis(2,1)

    psi_ghz = np.sqrt(K)*(np.cos(gamma)*tensor([basis(2,0)]*3) +np.sin(delta)*np.exp(gamma*1j)*tensor([phi_a,phi_b, phi_c]))
    return psi_ghz
```

```python
random_ghz = []
for i in range(1000):
    random_ghz.append(rand_ghz())
```

```python
aux = random_ghz[0]
```

```python
real = np.real(np.array(aux).flatten())
print(real)
imag = np.imag(np.array(aux).flatten())
print(imag)
vec = np.zeros(len(real)*2)
vec[::2] = real
vec[1::2] = imag
np.append(vec, 1)
```

```python
flatten_ghz = []
for i in random_ghz:
    aux = i
    vec = np.zeros(len(real)*2)
    vec[::2] = np.real(np.array(aux).flatten())
    vec[1::2] = np.imag(np.array(aux).flatten())
    flatten_ghz.append(np.append(vec, int(1)))

flatten_separable = []
for i in final_states:
    aux = i
    vec = np.zeros(len(real)*2)
    vec[::2] = np.real(np.array(aux).flatten())
    vec[1::2] = np.imag(np.array(aux).flatten())
    flatten_separable.append(np.append(vec, int(0)))

```
